```python
# import statements
from time import sleep
from json import dumps
from datetime import timezone
import time
import datetime
from kafka import KafkaProducer
import random
import datetime as dt
import csv
 
#reading csb file
def read_csv(fileName):
    '''Read the CSV file clickstream.csv'''
    '''you can use csv.DictReader'''
    data = []
    file_data = csv.DictReader(open(fileName))
    for row in file_data:
        data.append(row)
    
    return data

#function for publishing the message
def publish_message(producer_instance, topic_name, data):
    try:
        producer_instance.send(topic_name, value=data)
        print('Message published successfully. Data: ' + str(data))
    except Exception as ex:
        print('Exception in publishing message.')
        print(str(ex))
        
def connect_kafka_producer():
    _producer = None
    try:
        _producer = KafkaProducer(bootstrap_servers=['localhost:9092'],
                                  value_serializer=lambda x: dumps(x).encode('ascii'),
                                  api_version=(0, 10))
    except Exception as ex:
        print('Exception while connecting Kafka.')
        print(str(ex))
    finally:
        return _producer
    
    #function to change data types       
def correct_datatype(dic):
    for i in dic:
        i['sequence'] = int(i['sequence'])
        i['machine'] = int(i['machine'])
        i['PID'] = int(i['PID'])
        i['TRUN'] = int(i['TRUN'])
        i['TSLPI'] = int(i['TSLPI'])
        i['TSLPU'] = int(i['TSLPU'])
        i['POLI'] = str(i['POLI'])
        i['NICE'] = int(i['NICE'])
        i['PRI'] = int(i['PRI'])
        i['RTPR'] = int(i['RTPR'])
        i['CPUNR'] = int(i['CPUNR'])
        i['Status'] = str(i['Status'])
        i['EXC'] = int(i['EXC'])
        i['State'] = str(i['State'])
        i['CPU'] = float(i['CPU'])
        i['CMD'] = str(i['CMD'])
    return dic
        

        
if __name__ == '__main__':
   
    topic = 'process_final199'
    cRows = read_csv('Streaming_Linux_process.csv')
    cRows = correct_datatype(cRows)
    print('Publishing records..')
    producer4 = connect_kafka_producer()

    initial_no4 = 0
    initial_no5 = 0
    initial_no6 = 0
    initial_no7 = 0
    initial_no8 = 0
    
    while True:
        current_time = int(time.mktime(dt.datetime.utcnow().replace(microsecond=0).timetuple()))
        
        #generating random records for X with machne 4
        random_no4 = initial_no4 + random.randint(10,50)
        size = len([x for x in cRows if x['machine'] == 4])
        if random_no4 >= size - 1:
            diff = random_no4 - (size - 1)
            data4 = [x for x in cRows if x['machine'] == 4][initial_no4:size-1]
            new_list = [x for x in cRows if x['machine'] == 4][0:diff]
            for i in new_list:
                data4.append(i)
            
            initial_no4 = diff
        else:
            data4 =  [x for x in cRows if x['machine'] == 4][initial_no4:random_no4]
            initial_no4 = random_no4
       
        #generating random records for X with machne 5
        random_no5 = initial_no5 + random.randint(10,50)
        size = len([x for x in cRows if x['machine'] == 5])
        if random_no5 >= size - 1:
            diff = random_no5 - (size - 1)
            data5 = [x for x in cRows if x['machine'] == 5][initial_no5:size-1]
            new_list = [x for x in cRows if x['machine'] == 5][0:diff]
            for i in new_list:
                data5.append(i)
            
            initial_no5 = diff
        else:
            data5 =  [x for x in cRows if x['machine'] == 5][initial_no5:random_no5]
            initial_no5 = random_no5
        #appending dictionary to data4 for publishing
        for i in data5:
            data4.append(i)

  
        #generating random records for X with machne 6
        random_no6 = initial_no6 + random.randint(10,50)
        size = len([x for x in cRows if x['machine'] == 6])
        if random_no6 >= size - 1:
            diff = random_no6 - (size - 1)
            data6 = [x for x in cRows if x['machine'] == 6][initial_no6:size-1]
            new_list = [x for x in cRows if x['machine'] == 6][0:diff]
            for i in new_list:
                data6.append(i)
            initial_no6 = diff
        else:
            data6 =  [x for x in cRows if x['machine'] == 6][initial_no6:random_no6]
            initial_no6 = random_no6
        #appending dictionary to data4 for publishing
        for i in data6:
            data4.append(i)
        
        #generating random records for X with machne 7
        random_no7 = initial_no7 + random.randint(10,50)
        size = len([x for x in cRows if x['machine'] == 7])
        if random_no7 >= size - 1:
            diff = random_no7 - (size - 1)
            data7 = [x for x in cRows if x['machine'] == 7][initial_no7:size-1]
            new_list = [x for x in cRows if x['machine'] == 7][0:diff]
            for i in new_list:
                data7.append(i)
            
            initial_no7 = diff
        else:
            data7 =  [x for x in cRows if x['machine'] == 7][initial_no7:random_no7]
            initial_no7 = random_no7
        #appending dictionary to data4 for publishing
        for i in data7:
            data4.append(i)
        
        #generating random records for X with machne 8
        random_no8 = initial_no8 + random.randint(10,50)
        size = len([x for x in cRows if x['machine'] == 8])
        if random_no8 >= size - 1:
            diff = random_no8 - (size - 1)
            data8 = [x for x in cRows if x['machine'] == 8][initial_no8:size-1]
            new_list = [x for x in cRows if x['machine'] == 8][0:diff]
            for i in new_list:
                data8.append(i)
            
            initial_no8 = diff
        else:
            data8 =  [x for x in cRows if x['machine'] == 8][initial_no8:random_no8]
            initial_no8 = random_no8
        #appending dictionary to data4 for publishing
        for i in data8:
            data4.append(i)
            
        
        #adding timestamp to each dictionary
        for i in data4:
            i['ts'] = current_time
        
        
     
        
        #publishing the data finally
        publish_message(producer4, topic, data4)
        
        
        sleep(5)
```
